```python
# напишем функцию для визуального представления пропущенных значений по столбцам

def pass_value_barh(df):
    (
        (df.isna().mean()*100)
        .to_frame()
        .rename(columns = {0:'space'})
        .query('space > 0')
        .sort_values(by = 'space', ascending = True)
        .plot(kind= 'barh', figsize=(20,7), legend = False, fontsize = 16)
        .set_title('Доля пропущенных значений' + "\n", fontsize = 22, color = 'SteelBlue')    
    );    
```

```python
# применим функцию

pass_value_barh(data)
```

Заполним пропуски следующим образом:

'locality_name' (Наименования населенных пунктов) - названия придумать нельзя, но возможно заменить пропуски на 'Нет названия'.  


'is_apartment' (аппартаменты или нет) - если нет значений, значит пропуски можно менять на булевое значение 'False'.  


'balcony' (кол-во балконов) - если пропуск, значит балконов в квартире нет и можно заменить NAN на ноль.  


'parks_around3000' (кол-во парков) - если пропуск, значит парков рядом нет и можно заменить NAN на ноль.  


'ponds_around3000' (кол-во водоемов) - если пропуск, значит водоемов рядом нет и можно заменить NAN на ноль.

```python
# Заполним пустые значения путем логичной замены, где это возможно

data['locality_name'] = data['locality_name'].fillna('Нет названия')
data['is_apartment'] = data['is_apartment'].fillna(False)
data['balcony'] = data['balcony'].fillna(0)
data['parks_around3000'] = data['parks_around3000'].fillna(0)
data['ponds_around3000'] = data['ponds_around3000'].fillna(0)
```

```python
# отсортируем столбец 'ceiling_height' для проверки значений

data['ceiling_height'].sort_values(ascending=False)
```

Как мы видим, в данных имеются слишком высокие значения потолков в квартирах. Для того, чтобы исправить ошибку, необходимо сделать фильтр только по тем квартирам, которые имеют высоту потолков от 10 и выше,а затем разделить двузначные значения на 10, трехзначные - на 100.

```python
# сохраним в переменную высоту потолков от 10 м и выше

high_values = data[data['ceiling_height'] >= 10]

# Заменим высокие значения потолков на новые значения

for height in high_values['ceiling_height']:
    if height < 100:
        data['ceiling_height'] = data['ceiling_height'].replace(height, (height / 10.0))
    else:
        data['ceiling_height'] = data['ceiling_height'].replace(height,(height / 100.0))

data['ceiling_height'].unique()
```

Высота потолков в некоторых квартирах была слишком завышена. Можно предположить, что по ошибке неверно поставили разделитель.

```python
# Заменим пропуски в столбцах 'kitchen_area' и 'living_area' путем умножения общей площади на долю помещений

data.loc[data['kitchen_area'].isna(),'kitchen_area'] = data['total_area'] * data['share_kitchen_area']
data.loc[data['living_area'].isna(),'living_area'] = data['total_area'] * data['share_living_area'] 

# проверим количество пропусков

data.isna().sum()
```

```python
# Рассчитаем долю других площадей в составе общей, чтобы определить медианное значение
data['share_other_area'] = 1- data['share_area_sum']

# Выведем харакетристики по столбцу 'share_other_area'
data['share_other_area'].describe()
```

```python
# После заполнения пропусков, необходимо перепроверить значения в столбцах

# Напишем функцию, которая проверит меньше ли площадь остальных помещений, чем сумма площадей кухни и жилой площади

def check_columns(data):
    
    if data['total_area']*0.25 <= (data['living_area'] + data['kitchen_area']):
        return 'Все верно'
    elif data['total_area']*0.25 > (data['living_area'] + data['kitchen_area']):
        return 'Неверно!'

data['check_data'] = data.apply(check_columns, axis=1)

# Проверим есть ли ошибки

data.query('check_data == "Неверно!"')
```

Данные строки содержат в себе некорректные данные о площадях. Данные показывают, что суммарная доля кухни и жилой площади в составе общей площади намного меньше, чем доля нежилых помещений.

```python
# Сгруппируем данные по площади помещений и найдем медианные значения цен,
# затем построим диаграмму рассеяния для определения связи значений

grouped_total_area = data.pivot_table(index='total_area', values='last_price', aggfunc='median').reset_index()
grouped_living_area = data.pivot_table(index='living_area', values='last_price', aggfunc='median').reset_index()
grouped_kitchen_area = data.pivot_table(index='kitchen_area', values='last_price', aggfunc='median').reset_index()

fig, ax = plt.subplots(3, figsize=(17, 20))
ax[0].scatter(x = grouped_total_area['total_area'], y = grouped_total_area['last_price'], c = 'green')
ax[0].set_title('1. Взаимосвязь между общей площадью и ценой',fontsize=14)
ax[0].set_xlabel("Общая площадь, кв.м")
ax[0].set_ylabel("Стоимость квартир")

ax[1].scatter(x = grouped_living_area['living_area'], y = grouped_living_area['last_price'])
ax[1].set_title('2. Взаимосвязь между жилой площадью и ценой',fontsize=14)
ax[1].set_xlabel("Жилая площадь, кв.м")
ax[1].set_ylabel("Стоимость квартир")

ax[2].scatter(x = grouped_kitchen_area['kitchen_area'], y = grouped_kitchen_area['last_price'], c = 'fuchsia')
ax[2].set_title('3. Взаимосвязь между  площадью кухни и ценой',fontsize=14)
ax[2].set_xlabel("Площадь кухни, кв.м")
ax[2].set_ylabel("Стоимость квартир")
plt.show()
```
